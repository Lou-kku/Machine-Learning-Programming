```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Dense, Flatten, Dropout,
    Conv1D, MaxPooling1D, BatchNormalization,
    GlobalAveragePooling1D, Activation,
    add, LayerNormalization, MultiHeadAttention, Embedding
)
# Adam 옵티마이저를 직접 임포트하여 learning_rate를 설정합니다.
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import (
    EarlyStopping, ReduceLROnPlateau
)
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import warnings

# pandas 경고 무시
warnings.filterwarnings('ignore', category=FutureWarning)

# ### Cell 1: 데이터 로드 및 전처리 (Standardization 추가) ###

def load_file(filepath):
    dataframe = pd.read_csv(filepath, header=None, sep='\s+')
    return dataframe.values

def load_group(filenames, prefix=''):
    loaded = list()
    for name in filenames:
        data = load_file(prefix + name)
        loaded.append(data)
    loaded = np.dstack(loaded)
    return loaded

def load_dataset_group(group, prefix=''):
    filepath = prefix + group + '/Inertial Signals/'
    filenames = list()
    filenames += ['total_acc_x_'+group+'.txt', 'total_acc_y_'+group+'.txt', 'total_acc_z_'+group+'.txt']
    filenames += ['body_acc_x_'+group+'.txt', 'body_acc_y_'+group+'.txt', 'body_acc_z_'+group+'.txt']
    filenames += ['body_gyro_x_'+group+'.txt', 'body_gyro_y_'+group+'.txt', 'body_gyro_z_'+group+'.txt']
    X = load_group(filenames, filepath)
    y = load_file(prefix + group + '/y_'+group+'.txt')
    return X, y

def load_dataset(prefix=''):
    # Google Drive 경로
    trainX, trainy = load_dataset_group('train', prefix + '/content/drive/MyDrive/UCI HAR Dataset/')
    testX, testy = load_dataset_group('test', prefix + '/content/drive/MyDrive/UCI HAR Dataset/')

    trainy = trainy - 1
    testy = testy - 1

    trainy_one_hot = to_categorical(trainy)
    testy_one_hot = to_categorical(testy)

    print("--- 데이터 로드 완료 ---")
    return trainX, trainy, trainy_one_hot, testX, testy, testy_one_hot

# --- 데이터 로드 실행 ---
trainX, trainy, trainy_one_hot, testX, testy, testy_one_hot = load_dataset()

# --- 훈련/검증 데이터 순차 분리 (shuffle=False) ---
X_train, X_val, y_train_one_hot, y_val_one_hot, y_train, y_val = train_test_split(
    trainX, trainy_one_hot, trainy, test_size=0.2, random_state=100, shuffle=False
)
print("\n--- 훈련/검증 데이터 순차 분리 완료 (shuffle=False) ---")

# --- 데이터 표준화 (Standardization) ---
mu = X_train.mean(axis=(0,1), keepdims=True)
sigma = X_train.std(axis=(0,1), keepdims=True) + 1e-8
X_train = ((X_train - mu) / sigma).astype('float32')
X_val = ((X_val - mu) / sigma).astype('float32')
testX = ((testX - mu) / sigma).astype('float32')
print("\n--- 데이터 표준화 완료 (Train 통계 기준) ---")


# ### Cell 2: 변수 설정  ###
n_timesteps, n_features, n_outputs = X_train.shape[1], X_train.shape[2], y_train_one_hot.shape[1]
print(f"\n--- 변수 설정 완료 ---")
print(f"Timesteps: {n_timesteps}, Features: {n_features}, Outputs: {n_outputs}")


# ### Cell 3 : "Lighter" 트랜스포머 인코더 모델 정의 ###

def transformer_encoder_block(inputs, d_model, num_heads, ff_dim, dropout_rate=0.1):
    """
    트랜스포머 인코더 블록
    (Multi-Head Attention -> Add&Norm -> FeedForward -> Add&Norm)
    """
    # 1. Multi-Head Attention
    attn_input = LayerNormalization(epsilon=1e-6)(inputs)
    attention_output = MultiHeadAttention(
        num_heads=num_heads, key_dim=d_model // num_heads, dropout=dropout_rate
    )(attn_input, attn_input) # Self-Attention
    attn_output = Dropout(dropout_rate)(attention_output)
    # 1. Add & Norm (Residual Connection)
    out1 = add([inputs, attn_output])

    # 2. Feed Forward Network
    ffn_input = LayerNormalization(epsilon=1e-6)(out1)
    ffn_output = Dense(ff_dim, activation="relu")(ffn_input)
    ffn_output = Dropout(dropout_rate)(ffn_output)
    ffn_output = Dense(d_model)(ffn_output) # (d_model) 차원으로 다시 프로젝션
    return add([out1, ffn_output])

print("\n--- 'Lighter' 트랜스포머 모델 정의 ---")

# 하이퍼파라미터 (복잡도 감소)
d_model = 64        # 임베딩 차원 (유지)
num_heads = 2       # 어텐션 헤드 수 (4 -> 2)
ff_dim = 64         # 피드 포워드 네트워크 은닉층 (128 -> 64)
num_transformer_blocks = 2 # 트랜스포머 블록 수 (3 -> 2)
dropout_rate = 0.3  # 드롭아웃 비율 (0.25 -> 0.3)

inputs = Input(shape=(n_timesteps, n_features))

# 1. 피처 임베딩
x = Dense(d_model)(inputs)

# 2. 위치 임베딩
positions = tf.range(start=0, limit=n_timesteps, delta=1)
pos_embedding = Embedding(input_dim=n_timesteps, output_dim=d_model)(positions)
x = x + pos_embedding # 브로드캐스팅 덧셈
x = Dropout(dropout_rate)(x)

# 3. 트랜스포머 인코더 스택 (2개)
for _ in range(num_transformer_blocks):
    x = transformer_encoder_block(x, d_model, num_heads, ff_dim, dropout_rate)

# 4. 분류기 헤드 (규제 강화)
x = GlobalAveragePooling1D()(x)
x = Dropout(0.5)(x) # 0.4 -> 0.5
x = Dense(128, activation="relu")(x)
x = Dropout(0.5)(x) # 0.4 -> 0.5
outputs = Dense(n_outputs, activation="softmax")(x)

# 모델 생성
transformer_model_v2 = Model(inputs, outputs)

# !!! 수정: Optimizer의 Learning Rate를 0.0001로 낮춤 !!!
optimizer = Adam(learning_rate=1e-4)

# 모델 컴파일
transformer_model_v2.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

transformer_model_v2.summary()


# ### Cell 4: 훈련 (EarlyStopping 기준 = val_accuracy) ###
print("\n--- 모델 훈련 시작 (Lighter Transformer) ---")

# EarlyStopping 콜백 (patience=15 유지)
early_stopping_cb = EarlyStopping(
    monitor='val_accuracy',
    mode='max',
    patience=15,
    restore_best_weights=True
)

# 학습률 감소 (ReduceLROnPlateau) 콜백
reduce_lr_cb = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001)

# batch_size=32로 훈련
model_history = transformer_model_v2.fit(
    x=X_train, y=y_train_one_hot,
    epochs=50, # 최대 50 에포크
    batch_size=32,
    validation_data=(X_val, y_val_one_hot),
    callbacks=[early_stopping_cb, reduce_lr_cb],
    verbose=1
)
print("\n--- 모델 훈련 완료 ---")


# ### Cell 5: 훈련 및 검증 손실 시각화 ###
print("\n--- 훈련/검증 손실 시각화 ---")
loss = model_history.history['loss']
val_loss = model_history.history['val_loss']
epochs_range = range(1, len(loss) + 1)
plt.figure(figsize=(10, 6))
plt.plot(epochs_range, loss, 'y', label='Training loss')
plt.plot(epochs_range, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss (Lighter Transformer)')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()


# ### Cell 6: 훈련 및 검증 정확도 시각화 ###
print("\n--- 훈련/검증 정확도 시각화 ---")
acc = model_history.history['accuracy']
val_acc = model_history.history['val_accuracy']
plt.figure(figsize=(10, 6))
plt.plot(epochs_range, acc, 'y', label='Training acc')
plt.plot(epochs_range, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy (Lighter Transformer)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()


# ###  Cell 7: 모델 평가 ###
print("\n--- 최종 테스트 데이터 평가 시작 ---")
pred_probs = transformer_model_v2.predict(testX)
pred = np.argmax(pred_probs, axis=-1)

# Confusion Matrix 계산
cm = confusion_matrix(testy, pred)
print("\nConfusion Matrix:")
print(cm)

# Accuracy Score
acc_score = accuracy_score(testy, pred)
print(f"\nAccuracy Score: {acc_score:.4f}")

# Classification Report
print("\nClassification Report:")
print(classification_report(testy, pred, digits=3))

# Confusion Matrix 시각화
print("\n--- 혼동 행렬(Confusion Matrix) 시각화 ---")
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=range(6), yticklabels=range(6), cmap='Blues')
plt.title(f'Confusion Matrix (Accuracy: {acc_score:.4f})')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

print("\n--- 모든 과정이 완료되었습니다. ---")
```
